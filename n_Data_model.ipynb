```python
df_X = pd.read_csv('descriptors_output.csv')
```

```python
df_X = df_X.rename(columns={'Name':'CID'})
```

Now we can see we have a binary sequence for every compound. Those sequences are the PubChem fingerprints. More information about these fingerprints is provided in **'ML_model'** notebook.

```python
df_X
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

We'll replace outcome **Inactive** by 0 and **Active** by 1. This way, our machine learning model can understand the input. 

```python
df.Activity_Outcome[df.Activity_Outcome == 'Inactive'] = 0
df.Activity_Outcome[df.Activity_Outcome == 'Active'] = 1
```

```python
selecY = ['CID', 'Activity_Outcome']
df_Y = df[selecY]
```

```python
df_Y
```

Now we merge X and Y data in order to have one dataset which will be provided to the machine learning model. 

```python
df_model = pd.merge(df_X, df_Y, on='CID', how='inner')
```

```python
df_model.drop(columns='CID', inplace=True)
```

```python
df_model
```

```python
df_model.to_csv('Data_to_model.csv', index=False)
```
